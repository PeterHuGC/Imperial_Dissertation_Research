For years get the date by annual report date, and then get everything by boardid to gvkey and join details together.

```python
annual_report_details_df = df[['boardid', #'cikcode', # 'directorid',  # got rid of the direictor id
'timeretirement',
       'timerole', 'timebrd', 'timeinco', 'avgtimeothco', 'totnolstdbrd',
       'totnounlstdbrd', 'totnoothlstdbrd', 'totcurrnolstdbrd',
       'totcurrnounlstdbrd', 'totcurrnoothlstdbrd', 'noquals',
       'annualreportdate', 'remchgelast', 'recordid', 'salary',
       'bonus', 'other', 'penempcon', 'totalcompensation', 'valtoteqheld',
       'valltipheld', 'intvaloptheld', 'estvaloptheld', 'toteqlinkedwlth',
       'valeqaward', 'ltipvalue', 'intrvaloptaward', 'estvaloptaward',
       'toteqatrisk', 'totremperiod', 'bonusratio', 'eqlinkremratio',
       'wealthdelta', 'totaldirectcomp', 'perftotal', 'succession',
       'attrition', 'genderratio', 'nationalitymix', 'numberdirectors',
       'stdevtimebrd', 'stdevtimeinco', 'stdevtotnolstdbrd',
       'stdevtotcurrnolstdbrd', 'stdevnoquals', 'stdevage', 'networksize']]
```

```python
annual_report_details_df['annualreportdate'].isna().sum()
```

```python
annual_report_details_df = annual_report_details_df[(~annual_report_details_df['annualreportdate'].isna()) ]
annual_report_details_df.head()
```

```python
annual_report_details_df["annualreportdate"] = pd.to_datetime(annual_report_details_df['annualreportdate']) # get grant date details
annual_report_details_df["annualreport_year"] = annual_report_details_df["annualreportdate"].dt.year
```

```python
# get the boardex gvkey linking 

boardex_gvkey_linking_df = pd.read_csv("raw_data/boardex_gvkey_compustat_link_table.csv", index_col = 0)
```

```python
boardex_gvkey_linking_df
```

```python
boardex_gvkey_linking_df = boardex_gvkey_linking_df[boardex_gvkey_linking_df["preferred"]==1][["gvkey", "companyid"]].drop_duplicates()
```

```python
merged_stats = pd.merge(boardex_gvkey_linking_df, annual_report_details_df, left_on = "companyid", right_on = "boardid")

merged_stats.head()
```

```python
merged_stats.columns
```

```python
final_merged_stats = merged_stats[['gvkey', 'annualreport_year', 'timeretirement', 'timerole',
       'timebrd', 'timeinco', 'avgtimeothco', 'totnolstdbrd', 'totnounlstdbrd',
       'totnoothlstdbrd', 'totcurrnolstdbrd', 'totcurrnounlstdbrd',
       'totcurrnoothlstdbrd', 'noquals', 'annualreportdate', 'remchgelast',
       'recordid', 'salary', 'bonus', 'other', 'penempcon',
       'totalcompensation', 'valtoteqheld', 'valltipheld', 'intvaloptheld',
       'estvaloptheld', 'toteqlinkedwlth', 'valeqaward', 'ltipvalue',
       'intrvaloptaward', 'estvaloptaward', 'toteqatrisk', 'totremperiod',
       'bonusratio', 'eqlinkremratio', 'wealthdelta', 'totaldirectcomp',
       'perftotal', 'succession', 'attrition', 'genderratio', 'nationalitymix',
       'numberdirectors', 'stdevtimebrd', 'stdevtimeinco', 'stdevtotnolstdbrd',
       'stdevtotcurrnolstdbrd', 'stdevnoquals', 'stdevage', 'networksize']]
```

```python
final_merged_stats = final_merged_stats.groupby(["gvkey", "annualreport_year"]).mean().reset_index()
```
