```python
import pandas as pd
import numpy as np
```

```python
patent_metadata_df = pd.read_csv("raw_data/us_patent_all_us_pharma_gvkey_metadata.csv", index_col= 0)
```

```python
patent_metadata_df["date"] = pd.to_datetime(patent_metadata_df['link_bdate'])
patent_metadata_df["year"] = patent_metadata_df["date"].dt.year
```

```python
patent_metadata_df
```

```python
patent_metadata_df_filtered = patent_metadata_df[["gvkey", "gvkey_numeric", "year", "patnum"]] 
```

```python
patent_features_df = patent_metadata_df_filtered.groupby(["gvkey", "gvkey_numeric", "year"]).count().reset_index()

patent_features_df.head()
```

```python
patent_features_df.sort_values(["gvkey", "year"])
```

```python
patent_features_df.to_csv("features/company_patent_count.csv")
```

```python
patent_metadata_join_key_df = patent_metadata_df[["patnum", "gvkey_numeric", "gvkey"]]
patent_metadata_join_key_df
```

```python
patent_references_df = pd.read_csv("raw_data/us_patent_all_us_pharma_all_patent_details_need_gvkey.csv", index_col = 0)

patent_references_df.head()
```

```python
patent_references_df.shape
```

```python
patent_references_df = pd.merge(patent_metadata_join_key_df, patent_references_df, on = "patnum")

patent_references_df
```

```python
# get year , and get the cites for year and dates

patent_references_df["date"] = pd.to_datetime(patent_references_df['grantdate']) # get grant date details
patent_references_df["grant_year"] = patent_references_df["date"].dt.year
```

```python
patent_references_df.head()
```

```python
# link gvkey for results, get the forward and backward cites
```

```python
patent_details_df = patent_references_df[["gvkey", "patnum", "grant_year", "backward_cites", "forward_cites"]]

patent_details_df
```

```python
patent_details_df.sort_values(["gvkey", "patnum"])
```

```python
patent_details_stats_df = patent_details_df.groupby(["gvkey", "grant_year"]).agg({"patnum" : "count", "backward_cites" : "sum", "forward_cites" : "sum"}).reset_index()
patent_details_stats_df
```

```python
patent_details_stats_df.to_csv("features/patent_forward_backward_cites_2011_2019.csv") # cites are an even smaller subset - that's it!!

# left join where it's needed - not a full join for the data.
```
