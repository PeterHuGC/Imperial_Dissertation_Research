```python
import pandas as pd
import numpy as np
```

```python
original_df = pd.read_csv("data/boardex_director_employment_data_comp_net_2011_2023.csv", index_col=0)
original_df.head()
```

```python
original_df.shape
```

```python
seeded_lda_df = pd.read_csv("data/boardex_director_employment_data_comp_net_2011_2023_results_with_everything.csv")
seeded_lda_df.head()
```

```python
merged_df = pd.merge(original_df, seeded_lda_df, how = "left", on = ["rowtype",	"directorname" , "companyname", "brdposition", "rolename", "fulltextdescription", "ned", "directorid", "companyid", "datestartrole", "dateendrole"])

merged_df.head()
```

```python
final_merged_topics_df = merged_df.iloc[:, list(range(original_df.shape[1])) + [-1]]
final_merged_topics_df
```

```python
# fill final topic with other category

final_merged_topics_df["Topic"] = final_merged_topics_df["Topic"].fillna("other")

```

```python
final_merged_topics_df
```

```python
final_merged_topics_df["Topic"].value_counts()
```

```python
final_merged_topics_df.to_csv("data/boardex_director_employment_data_comp_net_2011_2023_final_csv_merged_category_merged_combined.csv")
```
